```python
# 1.3 and 1.4
# In question 1.3 we have to find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India.

import pandas as pd
import matplotlib.pyplot as plt
a=pd.read_csv("Downloads/zomato.csv",encoding='ISO-8859-1')
df=a.copy()
df.columns=['res_id','res_name','code','city','address','locality','loc_verb','longitude','latitude','cuisines','avg_cost',
           'currency','has_table','online','is_delivering','switch_to','price_range','agg_rating','rating_color','rating_text',
           'votes']
df2=df.copy()
def Check(x):
    if(x == 'New Delhi'):
        return 'NCR'
    elif (x == 'Delhi'):
        return 'NCR'
    elif (x=='Ghaziabad'):
        return 'NCR'
    elif(x == 'Faridabad'):
        return 'NCR'
    elif (x=='Gurgaon'):
        return 'NCR'
    elif(x=='Noida'):
        return 'NCR'
    return x
df2['city'] = df2['city'].apply(Check)
cuis=df2[df2.code==1]
rest=cuis[cuis.city!='NCR']
# we store the value of cuisines in rest where region is not NCR
d={}
for ele in rest.cuisines:
    p=str(ele)
    q=p.split(",")
    for e in q:
        e=e.strip()
        if e in d:
            d[e]+=1
        else:
            d[e]=1


ncr=cuis[cuis.city=='NCR']
d1={}
for ele in ncr.cuisines:
    p=str(ele)
    q=p.split(",")
    for e in q:
        e=e.strip()
        if e in d1:
            d1[e]+=1
        else:
            d1[e]=1


roi_name=[]
roi_values=[]
print("Rest Of India")
for i in range(10):
    ans=''
    max=0
    for j in d:
        if d[j]>max:
            max=d[j]
            ans=j
    print(ans,max)
    roi_name.append(ans)
    roi_values.append(max)
    del d[ans]   
plt.bar(roi_name,roi_values)
plt.xticks(rotation=90)
plt.title("Top Cuisines in Rest of India")
plt.xlabel("Name of Cuisines")
plt.ylabel("Number of Cuisines")
plt.show()

ncr_name=[]
ncr_values=[]
print("NCR")
for i in range(10):
    ans1=''
    max1=0
    for j in d1:
        if d1[j]>max1:
            max1=d1[j]
            ans1=j
    print(ans1,max1)
    ncr_name.append(ans1)
    ncr_values.append(max1)
    del d1[ans1]    
plt.bar(ncr_name,ncr_values)
plt.xticks(rotation=90)
plt.title("Top Cuisines in NCR")
plt.xlabel("Name of Cuisines")
plt.ylabel("Number of Cuisines")
plt.show()


# for question 1.4
print(" FOR 2.4")
print()

plt.bar(roi_name,roi_values)
plt.xticks(rotation=90)
plt.title("Top Cuisines in Rest of India")
plt.xlabel("Name of Cuisines")
plt.ylabel("Number of Cuisines")
plt.show()


plt.bar(ncr_name,ncr_values)
plt.xticks(rotation=90)
plt.title("Top Cuisines in NCR")
plt.xlabel("Name of Cuisines")
plt.ylabel("Number of Cuisines")
plt.show()

plt.scatter(ncr_name,ncr_values,color='blue',label='NCR') 
plt.scatter(ncr_name,roi_values,color='red',label = 'ROI')
plt.title("Difference ")
plt.xticks(rotation=90)
plt.xlabel('Cuisines') 
plt.ylabel('Number')
plt.legend()
plt.show()
```
