```python
# In this question we have to find how Number of cuisines affects the rating of the restaurant.
# for this we have to make a suitable graph to show this
import pandas as pd
import matplotlib.pyplot as plt

zomato = pd.read_csv("Downloads/zomato.csv",encoding = "ISO-8859-1")
df = zomato.copy()

rating = df["Aggregate rating"]
cuisine = df["Cuisines"]
cuisine = cuisine.apply(lambda x : str(x).split(","))
cuisine = cuisine.apply(lambda x : len(x)) 
df["No of Cuisines"] = cuisine
ans = df.groupby("No of Cuisines")["Aggregate rating"].mean()
ind = ans.index
val = ans.values
plt.bar(ind,val)
plt.title("Number of Cuisines vs Aggrerate Rating")
plt.ylabel("Rating")
plt.xlabel("Number of Cuisines")
plt.show()

```
