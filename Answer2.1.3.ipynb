```python
# In this question we have to find how Average cost of restaurant affects the rating of the restaurant.
# for this we have to make a suitable graph to show this
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

zomato = pd.read_csv("Downloads/zomato.csv",encoding = "ISO-8859-1")
df = zomato.copy()
df = df[df["Country Code"] == 1]
rating = df["Aggregate rating"]
cost = df["Average Cost for two"]
c = np.array(cost)

plt.subplots(figsize=(12,6))
plt.scatter(cost,rating,c=c)
plt.title("Rating Vs Average Cost of Restaurants")
plt.ylabel("Rating")
plt.xlabel("Average Cost of Two")

plt.xticks(rotation=90)
plt.grid()
plt.show()
```
