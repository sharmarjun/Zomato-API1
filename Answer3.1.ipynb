```python
# 3.1 Plot the bar graph top 15 restaurants have a maximum number of outlets.
import pandas as pd
import matplotlib.pyplot as plt
a=a=pd.read_csv("Downloads/zomato.csv",encoding='ISO-8859-1')
df=a.copy()
df.columns=['res_id','res_name','code','city','address','locality','loc_verb','longitude','latitude','cuisines','avg_cost',
           'currency','has_table','online','is_delivering','switch_to','price_range','agg_rating','rating_color','rating_text',
           'votes']
d={}
# I store the data of outlets in dictionary d and then found top 15 restaurants with maximum number of outlets.
for i in df.res_name:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
xaxis=[] # for restaurant name
yaxis=[] # for number of outlets
for i in range(15): # for top 15 
    max=0
    ans=''
    for j in d:
        if d[j]>max:
            max=d[j]
            ans=j
    print(ans,max)
    xaxis.append(ans)
    yaxis.append(max)
    del d[ans]
plt.bar(xaxis,yaxis)
plt.xticks(rotation=90)
plt.title("Top 15 Resturants Outlets")
plt.xlabel("Name of Restaurant")
plt.ylabel("Number of Outlets")
plt.show()
```
