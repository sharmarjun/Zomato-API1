```python
# 3.3 Plot the bar graph top 10 restaurants in the data with the highest number of votes
# In this problem we have to find top 10 restaurants with highest number of votes
import pandas as pd
import matplotlib.pyplot as plt
a=a=pd.read_csv("Downloads/zomato.csv",encoding='ISO-8859-1')
df=a.copy()
df.columns=['res_id','res_name','code','city','address','locality','loc_verb','longitude','latitude','cuisines','avg_cost',
           'currency','has_table','online','is_delivering','switch_to','price_range','agg_rating','rating_color','rating_text',
           'votes']
# I use groupby to extract top 10 restaurants with highest number of votes
ans= df.groupby("res_name")["votes"].mean().sort_values(ascending=False)[0:10]
restaurant=ans.index
number=ans.values
for i in range(10):
    print(restaurant[i],int(number[i]))
plt.bar(restaurant,number)
plt.xticks(rotation=90)
plt.title("Top 10 Resturants ")
plt.xlabel("Name of Restaurant")
plt.ylabel("Number of Votes")
plt.show()
```
